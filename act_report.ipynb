WeRateDogs is a Twitter account that rates people's dogs with humorous comments about the dog. These ratings always have a denominator of 10 and their numerators are typically greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." <br>

Analysing the WeRateDogs tweets revealed some interesting insights which are discussed presently. <br>

**1. A typical rating on WeRateDogs lies between 10/10 and 13/10** <br> 





The diagram above shows that a typical WeRateDogs rating lies between 10/10 and 13/10. <br>

**2. Golden retriever and Labrador retriever are among the most featured dog breeds on WeRateDogs** <br>

Golden retriever, Labrador retriever, Pembroke, Chihuahua and pug are the Top 5 most featured dogs on WeRateDogs. Their combined proportion is approximately  23% of the total featured dogs. <br>

 
**3. Most of the WeRateDogs’ tweets contain one dog image** <br>



The diagram above shows that most of the tweets contain one image. One-image tweets make up 85.8% of the total tweets, about 10% have two images, 3.1% of the tweets contain 3 images and 1.5% have four images. <br>


**4. Tweets that have 4 images have the highest number of retweets and Likes on WeRateDogs** <br>


<br>




From the diagrams above, we know that tweets on WeRateDogs that contain four images have  the highest average retweets and number of Likes, followed by tweets that contain two and three images. Tweets that contain one image have the lowest average retweet and Likes.
